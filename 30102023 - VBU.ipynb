```python
#Principal Component Analysis (Temel Bileşenler Analizi)
#Dimension Reduction (Boyut Azaltma)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
cancer = load_breast_cancer()
df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
df.head()
```

```python
df.shape
```

```python
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
```

```python
pca = PCA(n_components = 3)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
print("Before PCA : ", df.shape)
print("After PCA : ", x_pca.shape)
```

```python
plt.figure(figsize = (6,6))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c = cancer['target'])

plt.xlabel('1. Temel Bileşen')
plt.ylabel('2. Temel Bileşen')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
top_five_artists = df.value_counts("artist_name").head()
top_five_artists
```

```python
top_five_artists.plot.barh()
plt.show()
```

```python
top_twenty_loudest_track = df[['loudness', 'track_name', 'artist_name']].sort_values(by = 'loudness', ascending = False).head(20)
top_twenty_loudest_track
```

```python
plt.figure(figsize = (12,7))
sns.barplot(x = 'loudness', y = 'artist_name', data = top_twenty_loudest_track)
plt.title("Most Dancable Tracks")
plt.show()
```

```python
most_dancable_track = df[["danceability", "track_name", "artist_name"]].sort_values(by = 'danceability', ascending = False).head(5)
most_dancable_track
```

```python
plt.figure(figsize = (12, 7))
sns.barplot(x = "danceability", y = "track_name", data = most_dancable_track)
plt.title("Most Danceable Tracks")
plt.show()
```

```python
top_ten_inst_track = df[["instrumentalness", "track_name", "artist_name"]].sort_values(by = "instrumentalness", ascending = False).head(10)
top_ten_inst_track
```

```python
plt.figure(figsize = (12, 5))
labels = top_ten_inst_track['track_name']
plt.pie(x = 'instrumentalness', data = top_ten_inst_track, autopct = '%1.2f%%', labels = labels)
plt.title('Top 10 Inst Tracks')
plt.show()
```

```python
popularity_more_than_70 = df[df.popularity > 70].sort_values(by = "popularity", ascending = False)
popularity_more_than_70
```
